```python
input = open("input.txt").readlines()
input = [x.rstrip() for x in input]
```

```python
alg = input[0]
```

```python
import numpy as np
img = []
for line in input[2:]:
  img.append(np.array([c == '#' for c in line],dtype=int))
img = np.array(img)

for i in range(50):
  rows,cols = img.shape
  img = np.pad(img, pad_width=2, mode='constant', constant_values=1 if alg[0] == '#' and i%2 == 1 else 0)
  new_img = np.zeros((rows+2,cols+2),dtype=int)
  for row in range(rows+2):
    for col in range(cols+2):
      idx = int("".join(__builtin__.map(str,img[row:row+3,col:col+3].flatten().tolist())),2)
      new_img[row,col] = 1 if alg[idx] == '#' else 0
  img = new_img
  print(img, img.shape, np.count_nonzero(img))
```
