```python
def vi(d):
  return(ord(d)-ord('w'))
def isint(d):
    try:
        int(d)
        return True
    except ValueError:
        return False
def run(prog, num):
  vs = [0]*4
  ni = 0
  shrink = False
  for ln,line in enumerate(prog):
    i = line.split()
    if i[0] == "inp":
      print(ln//18, vs)
      vs[vi(i[1])] = int(num[ni])
      ni += 1
    else:
      b = int(i[2]) if isint(i[2]) else vs[vi(i[2])]
      if i[0] == "add":
        vs[vi(i[1])] = b + vs[vi(i[1])]
      elif i[0] == "mul":
        vs[vi(i[1])] = b * vs[vi(i[1])]
      elif i[0] == "div":
        shrink = b == 26
        vs[vi(i[1])] = vs[vi(i[1])] // b
      elif i[0] == "mod":
        vs[vi(i[1])] = vs[vi(i[1])] % b
      elif i[0] == "eql":
        if b != 0 and shrink: print("need to match",vs[vi(i[1])])
        vs[vi(i[1])] = 1 if b == vs[vi(i[1])] else 0
  return(vs)
```
